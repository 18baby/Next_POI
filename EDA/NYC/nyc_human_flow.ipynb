```python
import geohash2
from cartoframes.viz import *
from shapely.geometry import Polygon

def assign_time_period(hour):
    if 5 <= hour < 10:
        return 'morning'    # 출근 시간대
    elif 10 <= hour < 16:
        return 'afternoon'  # 일상 업무 시간대 
    elif 16 <= hour < 21:
        return 'evening'    # 퇴근 시간대 
    else:
        return 'night'

def assign_timezone(df):
    df['time_period'] = df['hour'].apply(assign_time_period)
    return df

def geohash_to_polygon(gh):
    """Geohash를 Polygon 영역으로 변환"""
    lat, lon, lat_err, lon_err = geohash2.decode_exactly(gh)
    lat_min, lat_max = lat - lat_err, lat + lat_err
    lon_min, lon_max = lon - lon_err, lon + lon_err
    return Polygon([
        (lon_min, lat_min), (lon_min, lat_max),
        (lon_max, lat_max), (lon_max, lat_min),
        (lon_min, lat_min)
    ])

def df_to_geohash_gdf(df, lat_col='latitude', lon_col='longitude', precision=6):
    """
    POI 위치를 geohash로 매핑하고 이를 GeoDataFrame으로 변환
    * precision=6 -> 약 1.2km x 0.6km 영역
    * precision=5 -> 약 2.5km x 1.2km 영역
    * precision=7 -> 약 152m x 152m 영역
    """
    # 1. geohash 인코딩
    df['geohash'] = df.apply(lambda row: geohash2.encode(row[lat_col], row[lon_col], precision=precision), axis=1)

    gh_stats = df.groupby('geohash').agg(
        avg_ridership=('avg_ridership', 'mean'),
        latitude=(lat_col, 'first'),
        longitude=(lon_col, 'first')
    )
     
    # 4. GeoDataFrame으로 변환
    gh_stats['geometry'] = gh_stats.index.map(geohash_to_polygon)
    gdf = gpd.GeoDataFrame(gh_stats.reset_index(), geometry='geometry', crs='EPSG:4326')

    return gdf
```

```python
nyc_trans = assign_timezone(nyc_trans)

nyc_trans_morning = nyc_trans[nyc_trans['time_period'] == 'morning']
nyc_trans_afternoon = nyc_trans[nyc_trans['time_period'] == 'afternoon']
nyc_trans_evening = nyc_trans[nyc_trans['time_period'] == 'evening']
nyc_trans_night = nyc_trans[nyc_trans['time_period'] == 'night']
```

```python
gdf_morning = df_to_geohash_gdf(nyc_trans_morning)
gdf_afternoon = df_to_geohash_gdf(nyc_trans_afternoon)
gdf_evening = df_to_geohash_gdf(nyc_trans_evening)
gdf_night = df_to_geohash_gdf(nyc_trans_night)
```

```python
import folium
from shapely.geometry import mapping
from folium import GeoJson, GeoJsonTooltip
import matplotlib.cm as cm
import matplotlib.colors as colors
from IPython.display import HTML
import branca.colormap as bcm
from matplotlib import colors as mcolors
from branca.element import Template, MacroElement

# 출력 크기 조정
def render_map_fixed_size(map_obj, width, height):
    map_html = map_obj._repr_html_()
    wrapped = f'''
    <div style="width: {width}; height: {height}; overflow: hidden; border:1px solid lightgray;">
        {map_html}
    </div>
    '''
    return HTML(wrapped)

def add_legend(m, colormap, vmin, vmax, caption="Count"):
    """folium 지도에 범례를 추가하는 함수"""

    # 브랑카 컬러맵 객체 생성
    color_scale = bcm.LinearColormap(
        colors=[colormap(i) for i in range(colormap.N)],
        vmin=vmin, vmax=vmax
    )
    color_scale.caption = caption
    color_scale.add_to(m)

# 각 변수별 수치값 시각화
def plot_geohash_continous(gdf, col='count', zoom=10.5, colormap_name='YlOrRd'):
    center = [gdf.geometry.centroid.y.mean(), gdf.geometry.centroid.x.mean()]
    
    # folium 지도 생성
    m = folium.Map(location=center, zoom_start=zoom, tiles='CartoDB positron')
    
    # 1. colormap 설정
    cmap = cm.get_cmap(colormap_name)
    norm = colors.Normalize(vmin=gdf[col].min(), vmax=gdf[col].max())

    # 2. count → 색상 변환 함수 정의
    def get_color(val):
        rgba = cmap(norm(val))      # RGBA → 0~1
        return colors.to_hex(rgba)    # → HEX로 변환 (#RRGGBB)
    
    # GeoJson 형태로 변환 및 스타일링
    folium.GeoJson(
        data=gdf,
        name='Geohash Polygons',
        style_function=lambda feature: {
            'fillColor': get_color(feature['properties'][col]),
            'color': 'gray',
            'weight': 0.5,
            'fillOpacity': 0.7,
            'opacity': 1.0
        },
        tooltip=GeoJsonTooltip(
            fields=['geohash', col],
            aliases=['Geohash:', f'{col.capitalize()}:'],
            sticky=False
        )
    ).add_to(m)
    
    # 범례 추가
    add_legend(m, cmap, gdf[col].min(), gdf[col].max(), caption=col.capitalize())
    display(render_map_fixed_size(m, width="1000px", height="600px"))
```

```python
plot_geohash_continous(gdf_morning, col='avg_ridership', zoom=10.5, colormap_name='YlOrRd')
plot_geohash_continous(gdf_afternoon, col='avg_ridership', zoom=10.5, colormap_name='YlOrRd')
plot_geohash_continous(gdf_evening, col='avg_ridership', zoom=10.5, colormap_name='YlOrRd')
plot_geohash_continous(gdf_night, col='avg_ridership', zoom=10.5, colormap_name='YlOrRd')
```
